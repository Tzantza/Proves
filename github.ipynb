## COMANDS

Terminal:

cd = change directory

mkdir "nom carpeta" = make directory

git init = creara dins la carpeta seleccionada un respositori nou Git

git commit = per comprovar que estem dins d'una carpeta git

clear = neteja la pantalla, pero no elimina re del que s'hagi fet anteriormnent a la sessi, es noms per despejar

git add "nom arxiu+extensio" = afegeix l'arxiu de la carpeta al repositori

git status = ensenya el que hi ha al respositori

git commit -m "Missatge" = Per posar un missatge que indica que ha canviat en el commit (checkpoint)

git add . = guarda TOTS els arxius de la carpeta a Git, sense haver de posar el nom un a un 

git checkout -b "nom branch, sense cometes= afegir un branch al masterbranch

git checkout main = torna a treballar al branch principal

git checkout "nom branc, sense cometes" = treballa al branch seleccionat

git merge main "nom branch creat" = fusiona el branch creat al main

git push -u origin main = per afegir arxius al main de github, posar el -u far que guardar el origin main pels seguents cops que posi git push, no calra posar a on ni res


Quan es modifica i guarda un arxiu, pero no es puja a git, al fer un git status surt en vermell "modified: nom arxiu"

Es com que informa de que no has guardat els canvis a git
